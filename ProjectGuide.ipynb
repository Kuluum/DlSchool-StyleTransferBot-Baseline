# Руководство по выполнению заключительного проекта DLSchool
Шаги, описанные ниже не являются обязательными и призываны помочь продвигаться вперед. Самое главное - резульат, поэтому если Вы чувствуете, что можете лучше, то вносите любые изменения.

## Основная часть
В основной части проекта Вам предлагается интегрировать медленный перенос стиля в телеграм бота.

1. Сначала посмотрите на бейзлайн и поймите, как устроен бот. Мы использовали библиотеку https://pypi.org/project/python-telegram-bot/ для написания самого бота. Она достаточно удобная и имеет хорошую документацию на гитхабе https://github.com/python-telegram-bot/python-telegram-bot. Изучите ее и позапускайте примеры.
2. Теперь к самому алгоритму переноса стиля. У нас есть посвященный ему семинар: https://github.com/python-telegram-bot/python-telegram-bot. Сам код можно найти в этоv репозитории https://github.com/DLSchool/deep_learning_2018-19/tree/master/14.%20GAN%20and%20Style%20transfer. Изучите материал по ссылкам и переходите к следующему шагу.
3. Теперь необходимо перенести код в класс, объявленный в бейзлайне, а потом протестировать его работу. Комментарии в коде помогут Вам. Для теста можно написать отдельный файл, который будет запускать трансфер стиля на каких-то картинках, хранящихся у Вас на компбютере.
4. Осталось дописать бота, чтобы он использовал класс с переносом стиля, и улучшить то, как бот взаимодействует с пользователем (написать красивые диалоги и кнопочки, предлагающие загрузить фото). Начать стоит с получения токена от BotFather и приветствия при подключении к боту (это позволит убедится, что все работает).

**(всего до 6 баллов)**

**За минимально работающую версию будет поставлен зачет (как синимум 3 балла).** 

5. (ОПЦИОНАЛЬНО) Скорее всего код, который Вы уже написали, работат только для одного пользователся: во время обработки картинки бот перестает работать и просто ждет конца работы нейросети. Это значит, что никакие сообщения, приходящие боту в такой момент не будут обрабатываться. Естественно, это можно исправить, но эта часть требует понимания того, как работает асинхронность. Мы не рассказывали про это, поэтому часть про параллельность не обязательная. Если Ваш код будет хорошо работать для многих пользователей, то Вы получите дополнительные 2 балла.

## Дополнительная часть
В этой части хочется добавить в наш бот более сложный алгоритм для обработки картинок. Вы можете выбрать что-то из статей, представленный ниже, или предложить что-то свое (тогда лучше написать мне и согласовать). Реализацию алгоритма не обязательно делать самому, но нужно чтобы она органично вписывалась в Ваш код. Это значит, что алгоритм нужно реализовать либо в виде обособленного модуля для бота, либо хорошо встроить в самого бота.
